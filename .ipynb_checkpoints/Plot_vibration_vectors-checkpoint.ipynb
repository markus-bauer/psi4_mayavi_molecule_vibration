```python
#---- functions to import molecule and vibration data --------

import sys
#from io import StringIO
from StringIO import StringIO
from collections import OrderedDict

# --- import vibrations ----
def read_vibration_file(vibfile):
    infile = open(vibfile,"r")
    content = infile.read()
    infile.close()
    freq_dict = OrderedDict({})
    freq = "NOTHING"
    #--- read file and make dict of blocks for each vibration
    for line in content.split("\n"):
        if "Frequency" in line:
            freq = float(line.split(" ")[-1])
            if freq not in freq_dict.keys() and freq != "NOTHING":
                freq_dict[freq] = ""
                block_start = 0
        if block_start == 1:
            freq_dict[freq] += line+"\n"
        if "mass" in line:
            block_start += 1

    freq_df_dict = OrderedDict({})
    for k,v in freq_dict.items():
        bufferdata=StringIO(v)
        if k not in freq_df_dict:
            freq_df_dict[k] = pd.read_csv(bufferdata,sep="\s*", skipinitialspace=True, names=["atom", "x", "y", "z", "mass"],
                                         comment="#")

    bufferdata=StringIO(vibfile)
    dfvib = pd.read_csv(bufferdata,sep="\s*", skipinitialspace=True, comment="#")
    
    for k,dfvib in freq_df_dict.items():
        for idx in dfvib.index:
            freq_df_dict[k].ix[idx,"vectorlength"] = vector_length(*dfvib.ix[idx,["x","y","z"]])
    
    return freq_df_dict

def read_molecule_file(molfile):
    return pd.read_csv(moleculefile,sep="\s*", header=None,names=["atom","x","y","z","mass"])



def vector_length(x,y,z):
    return np.sqrt(x**2 + y**2 + z**2)

```

```python
#--- mayavi version -- works only with python 2 ---

correction_factor = 0.91

common_colormap = "jet"
 
for k,dfvib in list(vibdict.items())[::-1]:
    # --- only for frequencies above 400
    if k <1170/0.91 and k>1150/0.91 :
        
        #fig =  mlab.figure(bgcolor=(1,1,1), fgcolor=(0,0,0))
        
        atom_coords = [np.array(dfmol.ix[:,"x"]),
                       np.array(dfmol.ix[:,"y"]),
                       np.array(dfmol.ix[:,"z"])]
        
        vib_coords = [np.array(dfvib.ix[:,"x"]),
                       np.array(dfvib.ix[:,"y"]),
                       np.array(dfvib.ix[:,"z"])]
        
        
        c = np.array(connections)
        x1,y1,z1 = (atom_coords[0][c[:,0]], 
                    atom_coords[1][c[:,0]], 
                    atom_coords[2][c[:,0]])

        x2,y2,z2 = (atom_coords[0][c[:,1]], 
                    atom_coords[1][c[:,1]], 
                    atom_coords[2][c[:,1]])
        
        colors = [atom_colors[dfmol.ix[_[0],"atom"]] for _ in connections]
        
        sticks = mlab.quiver3d(x1,y1,z1,
                               np.subtract(x2,x1),
                               np.subtract(y2,y1),
                               np.subtract(z2,z1),
                               color=(0,0,0),
                               line_width=3,
                               mode="2ddash",
                               scale_factor=1)
        #sticks.glyph.color_mode = "color_by_scalar"
        
        stksrc = sticks.mlab_source
#         for c in connections:
#             a1,a2 = c
#             x1,y1,z1 = df.ix[a1,["x","y","z"]]
#             x2,y2,z2 = df.ix[a2,["x","y","z"]]
#             color = atom_colors[df.ix[a1,"atom"]]
#             linecolor = "black"
#             stcks = mlab.plot3d([x1,x2],[y1,y2],[z1,z2], tube_radius=0.05, opacity=0.5, color=(1,1,1))
#             stksrc[c] = stcks.mlab_source

        #pointscalar = [atom_colors[_] for _ in df.ix[:,"atom"]]
        pointscalar = [relative_atomsize[_] for _ in dfmol.ix[:,"atom"]]
        

        pnts = mlab.points3d(atom_coords[0],atom_coords[1],atom_coords[2], pointscalar, colormap=common_colormap,
                             scale_mode="scalar",scale_factor=0.7)
        
        
        #---- plot atom index ---
#         for idx in range(np.shape(atom_coords)[1]-1):
#             mlab.text3d(atom_coords[0][idx],
#                         atom_coords[1][idx],
#                         atom_coords[2][idx],
#                         str(idx))
        #----------------------------
        #pnts = mlab.quiver3d(*atom_coords, mode="sphere", scale_factor=0.6)
        pnts.glyph.color_mode = "color_by_scalar"
        #pnts.module_manager.scalar_lut_manager.lut.table = colors
        pnts_src = pnts.mlab_source
        mlab.title(str(k)+"\n"+str(k*correction_factor), size=0.3)            
                
        @mlab.animate(delay=10)
        def anim():
            anim_strength= 1.5
            animcycle = np.linspace(0,anim_strength,12)
            animcycle = np.concatenate([animcycle,animcycle[::-1]])
#             repeats = 20
#            for n in range(repeats):
            while True: # endles repeats
                for aa in animcycle:
                
                    ax = np.add(atom_coords[0],np.multiply(vib_coords[0],aa))
                    ay = np.add(atom_coords[1],np.multiply(vib_coords[1],aa))
                    az = np.add(atom_coords[2],np.multiply(vib_coords[2],aa))
                    
                    pnts_src.set(x=ax,y=ay,z=az) 
                    
                    x1,y1,z1 = (ax[c[:,0]], ay[c[:,0]], az[c[:,0]])
                    x2,y2,z2 = (ax[c[:,1]], ay[c[:,1]], az[c[:,1]])
                    
                    x2 = np.subtract(x2,x1)
                    y2 = np.subtract(y2,y1)
                    z2 = np.subtract(z2,z1)

                    stksrc.set(x=x1,y=y1,z=z1, u=x2,v=y2,w=z2)
                    
                    yield
        
        def plot_ghosts():
            global connections
            anim_strength= 2
            nanimsteps = 5
            opacitysteps = 1.0/nanimsteps
            animcycle = np.linspace(0,anim_strength,nanimsteps)
            
            for aa in animcycle:
                
                for direction in [1,-1]:
                    ax = atom_coords[0]
                    ay = atom_coords[1]
                    az = atom_coords[2]
                    ax = np.add(ax,np.multiply(direction*vib_coords[0],aa))
                    ay = np.add(ay,np.multiply(direction*vib_coords[1],aa))
                    az = np.add(az,np.multiply(direction*vib_coords[2],aa))

                    
                    mlab.points3d(ax,ay,az, pointscalar, colormap=common_colormap, 
                                  scale_mode="scalar",scale_factor=0.7, opacity=opacitysteps)

                    x1,y1,z1 = (ax[c[:,0]], ay[c[:,0]], az[c[:,0]])
                    x2,y2,z2 = (ax[c[:,1]], ay[c[:,1]], az[c[:,1]])

                    x2 = np.subtract(x2,x1)
                    y2 = np.subtract(y2,y1)
                    z2 = np.subtract(z2,z1)

                    if direction == 1:
                        color=(1,0,0)
                    elif direction == -1:
                        color=(0,1,0)
                        
                    mlab.quiver3d(x1,y1,z1,x2,y2,z2,
                                  color=color,
                                  line_width=3,
                                  mode="2ddash",
                                  scale_factor=1,
                                  opacity=opacitysteps)
                    
            
            con = np.array(connections)
            x1,y1,z1 = (atom_coords[0][con[:,0]], 
                        atom_coords[1][con[:,0]], 
                        atom_coords[2][con[:,0]])
            
            x2,y2,z2 = (atom_coords[0][con[:,1]], 
                        atom_coords[1][con[:,1]], 
                        atom_coords[2][con[:,1]])

            sticks = mlab.quiver3d(x1,y1,z1,
                                   np.subtract(x2,x1),
                                   np.subtract(y2,y1),
                                   np.subtract(z2,z1),
                                   color=(0,0,0),
                                   line_width=4,
                                   mode="2ddash",
                                   scale_factor=1)
            
        anim()
        #plot_ghosts()
        mlab.show()
                
        #plt.savefig(filename_head+"{:.2f}.pdf".format(k))
        #mlab.savefig(filename_head+"{:.2f}.png".format(k), magnification=5)
```
